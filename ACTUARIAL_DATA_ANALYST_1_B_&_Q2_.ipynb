QUESTION 2 ZAMBIA DATA


```python
#loading the Zambia data
zambia_data = pd.read_excel('/content/Zambia Historical Payouts.xlsx', sheet_name="Payouts(all sites)")

# Display the first few rows
zambia_data.head()
```

```python
import matplotlib.pyplot as plt

# Rename the first column to 'Year'
zambia_data.rename(columns={zambia_data.columns[0]: "Year"}, inplace=True)

```

```python
# Compute average crop loss per year across all locations
zambia_data["Average_Loss"] = zambia_data.iloc[:, 1:].mean(axis=1)

```

```python
# Plot the trend of average crop loss over the years
plt.figure(figsize=(12, 6))
plt.plot(zambia_data["Year"], zambia_data["Average_Loss"], marker='o', linestyle='-', color='b', label="Avg Crop Loss")

```

Key Insights from the Analysis:
1. Crop losses have varied significantly from 1983 to 2023, with some years experiencing sharp spikes.
2. Certain years show higher-than-average losses, which could be linked to extreme weather events like droughts or floods.
Recent Trends (2000s–2020s):
3. There appears to be an increasing trend in crop losses over time, potentially indicating climate change impacts or increased agricultural risks.
