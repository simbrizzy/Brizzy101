```python
# importing libraries to help in web scrapping
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
```

```python
base_url = "https://www.airlinequality.com/airline-reviews/british-airways"
pages = 10
page_size = 100

reviews = []

# for i in range(1, pages + 1):
for i in range(1, pages + 1):

    print(f"Scraping page {i}")

    # Create URL to collect links from paginated data
    url = f"{base_url}/page/{i}/?sortby=post_date%3ADesc&pagesize={page_size}"

    # Collect HTML data from this page
    response = requests.get(url)

    # Parse content
    content = response.content
    parsed_content = BeautifulSoup(content, 'html.parser')
    for para in parsed_content.find_all("div", {"class": "text_content"}):
        reviews.append(para.get_text())

    print(f"   ---> {len(reviews)} total reviews")
```

```python
#creating a dataframe of the scraped data
df = pd.DataFrame()
df["reviews"] = reviews
df.head()
```

```python
import os

# Specify the directory path
new_directory = r"C:\Users\bryan\Desktop\FORAGE"

# Create the directory if it doesn't exist
os.makedirs(new_directory, exist_ok=True)

# Specify the file path in the new directory
new_file_path = os.path.join(new_directory, "BA_reviews.csv")

# Assuming df is your DataFrame
try:
    df.to_csv(new_file_path)
    print("Data saved successfully to:", new_file_path)
except Exception as e:
    print("An error occurred:", str(e))
```
